### Passagiers klasse analyse
Voor elke mogelijke passagiers klasse, toon het aantal overlevenden en sterfgevallen.
Zorg ook voor een bar plot.

```python
# Analyse op basis van P class 1
first_class = df_train[df_train['Pclass'] == 1]
first_class_count = len(first_class)
first_class_survivor = first_class[first_class['Survived'] == 1]
first_class_survivor_count = len(first_class_survivor)
first_class_dead_count = first_class_count - first_class_survivor_count
print("First class survivor count: ",first_class_survivor_count)
print("First class death count: ",first_class_dead_count)

# Analyse op basis van P class 2
second_class = df_train[df_train['Pclass'] == 2]
second_class_count = len(second_class)
second_class_survivor = second_class[second_class['Survived'] == 1]
second_class_survivor_count = len(second_class_survivor)
second_class_dead_count = second_class_count - second_class_survivor_count
print("Second class survivor count: ",second_class_survivor_count)
print("Second class death count: ",second_class_dead_count)

# Analyse op basis van P class 3
third_class = df_train[df_train['Pclass'] == 3]
third_class_count = len(third_class)
third_class_survivor = third_class[third_class['Survived'] == 1]
third_class_survivor_count = len(third_class_survivor)
third_class_dead_count = third_class_count - third_class_survivor_count
print("Third class survivor count:",third_class_survivor_count)
print("Third class death  count:",third_class_dead_count)

# Bar plot
plt.bar("1th Class Survivors", first_class_survivor_count)
plt.bar("1th Class Deaths", first_class_dead_count)
plt.bar("2th Class Survivors", second_class_survivor_count)
plt.bar("2th Class Deaths", second_class_dead_count)
plt.bar("3th Class Survivors", third_class_survivor_count)
plt.bar("3th Class Deaths", third_class_dead_count)
plt.title("All passengers")
plt.xlabel("Passengers outcome")
plt.ylabel("Amount of passengers")
plt.xticks(rotation=45)
plt.show()
```

### Overlevingskans berekenen
Met de code hieronder berekenen we de overlevingskans van een passagier met de attributen
van de passagier zoals: klasse, geslacht en leeftijd.
Dit wordt berekend door de passagiers met dezelfde attributen uit de 'train.csv' te filteren.
Vervolgens kunnen we het overlevingskans berekenen in percentages door naar de verhouding
te kijken van het aantal overlevenden en sterfgevallen.
De overlevingskans wordt gebruikt in het algoritme.

```python
# df_train_XXX = c: class, s: sex, a: age
df_train_c = df_train[df_train['Pclass'] == 1]
df_train_cs = df_train_c[df_train_c['Sex'] == 'male']
df_train_csa = df_train_cs[df_train_cs["Age"] <= 18]
passenger_count = len(df_train_csa) / len(df_train) * 100
survival_chance = len(df_train_csa[df_train_csa['Survived'] == 1]) / len(df_train_csa) * 100
print(survival_chance, "%")
```

### Algoritme
Met de code hieronder berekenen we of de passagier het Titanic gebeurtenis zal overleven.
Het algoritme filtert per passagier op zijn of haar klasse, leeftijd en geslacht.
Met de code hierboven hebben we de overlevingskans per passagier handmatig berekent.
Vervolgens pakken we een willekeurig overlevingskans en vergelijken we het met elkaar.
Als de passagiers overlevingskans onder de willekeurige overlevingskans ligt,
dan overleefd de passagier het.
Als de passagiers overlevingskans boven de willekeurige overlevingskans ligt,
dan sterft de passagier.

```python
# Kolom 'Survived' toevoegen met de standaard waarde van 0
df_test['Survived'] = 0

# Iterate over het dataframe 'df_test' waarbij we de index en het passenger waarde ophalen
for index, pas in df_test.iterrows() :
    # Maak een dict voor de passenger class
    pas_class = pas['Pclass']
    # Maak een dict voor de passenger age
    pas_age = pas["Age"]
    # Maak een dict voor de passenger sex
    pas_sex = pas["Sex"]
    
    # Bereken de overlevingskans van de passenger in percentage
    survival_chance = random.randint(1, 100)

    # Algoritme waarbij de passenger overlevingskans wordt berekend
    if pas_class == 1:
        if pas_age <= 18:
            if pas_sex == 'male' :
                if survival_chance <= 80 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 91 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 30 :
            if pas_sex == 'male' :
                if survival_chance <= 50 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 94 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 50 :
            if pas_sex == 'male' :
                if survival_chance <= 47 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 96 :
                    df_test.at[index, 'Survived'] = 1
        else:
            if pas_sex == 'male' :
                if survival_chance <= 19 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 100 :
                    df_test.at[index, 'Survived'] = 1
    elif pas_class == 2:
        if pas_age <= 18:
            if pas_sex == 'male' :
                if survival_chance <= 60 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 100 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 30 :
            if pas_sex == 'male' :
                if survival_chance <= 19 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 94 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 50 :
            if pas_sex == 'male' :
                if survival_chance <= 16 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 93 :
                    df_test.at[index, 'Survived'] = 1
        else:
            if pas_sex == 'male' :
                if survival_chance <= 8 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 67 :
                    df_test.at[index, 'Survived'] = 1
    else:
        if pas_age <= 18:
            if pas_sex == 'male' :
                if survival_chance <= 22 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 51 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 30 :
            if pas_sex == 'male' :
                if survival_chance <= 17 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 51 :
                    df_test.at[index, 'Survived'] = 1
        elif pas_age <= 50 :
            if pas_sex == 'male' :
                if survival_chance <= 16 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 46 :
                    df_test.at[index, 'Survived'] = 1
        else:
            if pas_sex == 'male' :
                if survival_chance <= 0 :
                    df_test.at[index, 'Survived'] = 1
            else :
                if survival_chance <= 100 :
                    df_test.at[index, 'Survived'] = 1

# Maak een dict aan van de dataframe 'df_test' waarbij alleen de kolommen 'PassengerId' en 'Survived' worden getoond
df_final = df_test[['PassengerId', 'Survived']]
# Bestandsnaam voor de df_final data
df_final_file_name = 'titanic_test_kaggle_submission.csv'
# Maak een csv file van de dict hierboven, waarbij de index niet wordt getoond
df_final.to_csv(df_final_file_name, index=False)
df_final
```
